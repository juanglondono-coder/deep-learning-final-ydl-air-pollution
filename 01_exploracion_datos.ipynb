1. Convertir Date-time a fecha y hora

```python
df['Datetime'] = pd.to_datetime(df['Datetime'])
df = df.sort_values('Datetime')
```

2. Definir variable respuesta y variables explicatorias

```python
target_col = 'NMHC(GT)'

feature_cols = [c for c in df.columns if c not in ['Datetime', target_col]]
```

3. Transformar varoles faltantes

```python
cols_with_minus200 = ['CO(GT)', 'NMHC(GT)', 'NOx(GT)', 'NO2(GT)']  # puedes revisar cuáles tienen -200

for c in cols_with_minus200:
    df[c] = df[c].replace(-200, np.nan)
```
